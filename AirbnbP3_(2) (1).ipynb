```python
#check if there missing data or null
# create a fcuntion to find and display NAs
def count_NAs(dfna, perc):
    NaList = []
    numOfRows = len(dfna)
    for i in dfna.columns:
        numOfNAs = dfna[i].isna().sum()
        print(f'{i:<30}, Rows: {numOfRows:>5}, NAs: {numOfNAs:<25}, precentage: {round(numOfNAs/numOfRows * 100.00,2)}')
        if ((numOfNAs/numOfRows) * 100.000) >= perc:
           NaList.append(i)
    print(f'columns with NAs over {perc}%: {NaList}')
    return NaList
# Display NAs
cutoff = 40
NAs = count_NAs(listings, cutoff)
print(f"Total with NAs over {cutoff}% column number :",len(NAs))
```

# Task3:**Exploratory Data Analysis**




# **univariate analysis **

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("darkgrid")
sns.kdeplot(data=listings, x="price", shade=True, color="green")
plt.xlabel('Price')
plt.ylabel('Density')
plt.title('Kernel Density Plot: Price Distribution')
plt.xlim(0, 1000)
plt.show()
```

```python

# Calculate mean price
mean_price = listings['price'].mean()

# Plot mean price on a bar graph
plt.bar('Mean Price', mean_price)
plt.xlabel('Price')
plt.ylabel('Mean Price')
plt.title('Mean Price of Listings')
plt.show()


# Compare 'price' using mean
mean_price = listings['price'].mean()
print(f"Mean Price: {mean_price}")
```

```python
# Take a sample of the 'neighbourhood' variable
sample_size = 100  # Specify the desired sample size
neighbourhood_sample = listings['neighbourhood'].sample(n=sample_size, random_state=42)

# Plot the sampled neighbourhood data
plt.figure(figsize=(15, 6))
plt.hist(neighbourhood_sample, bins=20, color='skyblue')
plt.xlabel('Neighbourhood')
plt.ylabel('Frequency')
plt.title('Distribution of Sampled Neighbourhood')
plt.grid(True)

# Reduce the number of x-axis labels
plt.xticks(rotation=90, ha='right')
plt.gca().set_xticklabels(neighbourhood_sample.unique()[::])  # Display every 5th unique neighbourhood label

plt.show()
```

```python
sns.boxplot(data = listings, x = "accommodates")
```

```python

plt.figure(figsize=(12, 6))

# Box plot for 'minimum_nights'
plt.subplot(1, 2, 1)
plt.boxplot(listings['minimum_nights'])
plt.ylabel('Minimum Nights')
plt.title('Distribution of Minimum Nights')

# Box plot for 'maximum_nights'
plt.subplot(1, 2, 2)
plt.boxplot(listings['maximum_nights'])
plt.ylabel('Maximum Nights')
plt.title('Distribution of Maximum Nights')

plt.tight_layout()
plt.show()
```

# **relationships between variables **

```python
# Take a sample of the data for analysis
sample = listings.sample(n=1000, random_state=42)

# Group the data by property type and calculate the average price
average_price_by_property = sample.groupby('property_type')['price'].mean().reset_index()

# Sort the data by average price in descending order
average_price_by_property = average_price_by_property.sort_values('price', ascending=False)

# Create the horizontal bar plot
plt.figure(figsize=(10, 8))
sns.barplot(x='price', y='property_type', data=average_price_by_property)
plt.xlabel('Price')
plt.ylabel('Property Type')
plt.title('Average Price by Property Type')
plt.grid(True)
plt.show()
```

```python
# Take a sample of the data for analysis
sample = listings.sample(n=500, random_state=42)

# Bar chart: price vs. neighborhood
average_price_by_neighborhood = sample.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
plt.figure(figsize=(25, 6))
plt.bar(average_price_by_neighborhood.index, average_price_by_neighborhood.values, color='skyblue')
plt.xlabel('Neighborhood')
plt.ylabel('Average Price')
plt.title('Price by Neighborhood')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()



```

# Visualize the data

```python
from wordcloud import WordCloud

neighbourhood_counts = listings['neighbourhood'].value_counts()

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(neighbourhood_counts)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Neighbourhood Frequency')
plt.show()
```

```python
# Bar plot for categorical variables
plt.figure(figsize=(10, 6))
listings['property_type'].value_counts().plot(kind='bar')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.title('Frequency of Property Types')
plt.xticks(rotation=90)
plt.show()
```

```python
# Line plot for time-based variables
listings['listingsDate'] = pd.to_datetime(listings['listingsDate'])
listings['reviewsDate'] = pd.to_datetime(listings['reviewsDate'])

plt.figure(figsize=(10, 6))
listings.groupby(listings['listingsDate'].dt.year)['listing_id'].count().plot(marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Listings')
plt.title('Number of Listings Over Time')
plt.show()
```


# Conclusion

Based on the exploratory data analysis (EDA) conducted, the following findings, insights, and recommendations can be summarized:

**Findings:**

-The distribution of listing prices showed that the majority of prices fell within the range of 0 to 1000, indicating a concentration of lower-priced listings.

-The price trend exhibited a positive skew, suggesting a higher concentration of lower-priced listings.

-The distribution of sampled neighborhoods provided insights into the frequency distribution of listings across different neighborhoods.

-The boxplot analysis of accommodates revealed the range and distribution of the variable, indicating the number of people the listing can accommodate.

-The boxplots for minimum and maximum nights showcased the allowed range of stay durations for the listings.

-Bar plots comparing average prices by property type and price by neighborhood highlighted variations in prices across different property types and neighborhoods.

-The line plot analysis of reviews over time revealed temporal trends in guest feedback.

 **Insights and Patterns:**

-There is a wide range of listing prices, with a concentration of lower-priced listings.

-Property type and neighborhood significantly impact listing prices, with certain property types and neighborhoods commanding higher average prices.

-The distribution of accommodates, minimum nights, and maximum nights provides insights into the range and variability of these variables.

-Reviews exhibit temporal trends, indicating potential seasonality or other factors affecting guest feedback.


 **Recommendations:***

-Price Setting: Property owners should consider the property type and neighborhood when setting prices. They can adjust their pricing strategy to be competitive within their specific property type and neighborhood, taking into account the observed price patterns.

-Guest Satisfaction: Property owners and managers should monitor and respond to reviews to ensure guest satisfaction.

- Addressing guest feedback promptly and proactively can enhance the overall guest experience and potentially lead to positive reviews and higher demand.

-Further Analysis: Additional research and analysis can be conducted to explore other variables that may influence listing prices, such as amenities, location proximity to attractions or transportation, and quality indicators.

- These variables can provide further insights into factors that drive pricing and help property owners optimize their listings.

-Market Positioning: Property owners can leverage the insights gained from the analysis to position their listings strategically within the Airbnb marketplace. By understanding the demand patterns and pricing dynamics of their property type and neighborhood, they can optimize their marketing efforts and attract the target audience effectively.
